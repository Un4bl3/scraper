```python
import requests as req
from bs4 import BeautifulSoup as bs
import pandas as pd


busena = req.get("https://www.coindesk.com/data/")

print(busena)

turinys = busena.content
bs()
strukturizuotas_html = bs(turinys, "lxml")
coins_names = []
coins_prices_INT = []
coins_prices_PERC_change = []
with open("Prices.html", 'wb') as file:
    file.write(strukturizuotas_html.prettify('utf-8'))
file.close()

for coins in strukturizuotas_html.find_all('span', {'class':'coin-name'}):
    coins_names.append(coins.text)
for coins in strukturizuotas_html.find_all('span', {'class':'price-values'})[::2]:
    coins_prices_INT.append(coins.text)
for coins in strukturizuotas_html.find_all('span', {'class':'price-values'})[1::2]:
    coins_prices_PERC_change.append(coins.text)

duom = pd.DataFrame(list(zip(coins_names,coins_prices_INT,coins_prices_PERC_change)), columns=['name','price','price change'])

duom.to_excel('Scraperio_rezultatas2.xlsx', index=False)
duom
#strukturizuotas_html.find_all('span', {'class':'price-values'})    
#coins_names
#coins_prices_INT
#coins_prices_PERC_change
```
